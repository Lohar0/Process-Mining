```python
import pandas as pd
import numpy as np
import os
import graphviz
import pm4py
```

```python
log = pm4py.read_xes("BPI Challenge 2017.xes")
```

```python
df = pm4py.convert_to_dataframe(log) 
df
```

```python
from pm4py.objects.log.util import interval_lifecycle
enriched_log = interval_lifecycle.assign_lead_cycle_time(log)
pd.DataFrame(enriched_log[0]).set_index('concept:name')
```

```python
# finding maximum number of traces

n = 10000
maximum = 0

for i in range(n): # maximum still returns 54 over the entire log
    df = pd.DataFrame(enriched_log[i]).set_index('concept:name')
    values = len(np.array(df['@@approx_bh_this_wasted_time']))
    if values > maximum:
        maximum = values

maximum
```

```python
complete_trace = np.zeros(maximum) # this should have the length of the longest trace
n = 100 # this should be equal to the number of applications
number_of_values = np.zeros(maximum) # this should have the length of the longest trace


for i in range(n):
    df = pd.DataFrame(enriched_log[i]).set_index('concept:name')
    values = np.array(df['@@approx_bh_this_wasted_time'])
    #array += np.array(df['@@approx_bh_this_wasted_time'])
    for j in range(len(values)):
        complete_trace[j] += values[j]
        number_of_values[j] += 1
    

average_time_per_state = np.around(complete_trace/number_of_values, decimals = 2)
average_time_per_state
```

```python
#''' giving the most occuring state for each position '''

import numpy as np
from collections import Counter

n = 100 # this should be equal to the number of applications
counts = {}


for i in range(n):
    df = pd.DataFrame(enriched_log[i]).set_index('concept:name')
    values = np.array(df.index)
    #array += np.array(df['@@approx_bh_this_wasted_time'])
    #print(values)
    for j in range(len(values)):
        if j in counts:
            counts[j].append(values[j])
        else:
            counts[j] = [values[j]]

for counts in counts.values():
    most_common = max(counts, key = counts.count)
    print(most_common)

#states

    
```
